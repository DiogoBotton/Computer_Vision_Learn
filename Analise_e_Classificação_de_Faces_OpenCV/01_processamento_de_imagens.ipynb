### Realizando processamentos em uma imagem

```python
# Importando a imagem
img = cv2.imread('images/duas_pessoas_sol.jpg')

# Plotando a imagem
plt.imshow(img)
```

Há uma distorção nas cores das imagens.

Isto acontece pois o OpenCV por padrão abre as imagens no formato BGR (Blue, Green e Red) e o matplotlib no padrão RGB (Red, Green e Blue). Então antes de exibir a imagem, precisamos fazer uma conversão.

```python
# Realizando a conversão de BGR para RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

Quando queremos identificar alguns padrões da imagem onde a cor não é tão importante, podemos converter a imagem para **escala de cinza** pois isto simplifica a análise, inclusive com isto, também é possível **executar classificadores em cascata em tempo real**, por ser muito veloz, por exemplo.

Classificadores em cascata são algoritmos que identificam por exemplo, o corpo inteiro de uma pessoa, em seguida, o rosto, o nariz, olhos, etc, em sequência.

```python
# Contem 3 canais: RGB
img.shape
```

```python
# Realizando a conversão para escala de cinza
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Informando para o pyplot com cmap que a imagem está em escala de cinza
plt.imshow(img_gray, cmap='gray')
```

Para deixar a imagem em escala de cinza, para cada pixel é realizado um cálculo básico, onde:

- Para cada pixel é somado os 3 canais (RGB) e dividido por 3.

Cada imagem basicamente é uma **matriz tri-dimensional** onde as duas primeiras dimensões é a resolução (comprimento e altura) e a terceira os canais RGB, que contem a informação das cores.

Na imagem em escala de cinza, temos somente o comprimento e altura, sem canais de cores (RGB).

Inclusive, o RGB é um número que vai de 0 até 255.
